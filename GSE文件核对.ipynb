```python
#遍历本地文件系统：遍历指定的本地目录，收集所有文件的路径、文件名和大小。
def human_readable_size(size):
    """将文件大小转换为人类可读的格式"""
    # 定义大小单位
    KB = 1024
    MB = KB * 1024
    GB = MB * 1024

    if size < MB:
        return f"{size/KB:.1f}KB"
    elif size < GB:
        return f"{size/MB:.1f}MB"
    else:
        return f"{size/GB:.1f}GB"

def list_local_files(directory):
    """遍历本地目录，返回文件信息列表"""
    file_info = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            path = os.path.join(root, file)
            size = os.path.getsize(path)
            readable_size = human_readable_size(size)
            file_info.append((path, readable_size))
    return file_info
```

```python
import pandas as pd

def split_file_info(path):
    """拆分文件路径信息"""
    parts = path.split('/')
    # 假设GSE编号总是路径的第四个部分（从1开始计数）
    gse_number = parts[-3]
    # 子目录名是第五个部分
    subdir_name = parts[-2]
    # 文件名是路径的最后一个部分
    file_name = parts[-1]
    return gse_number, subdir_name, file_name

def create_dataframe(file_info_list):
    """将文件信息列表转换为pandas DataFrame"""
    # 创建空的DataFrame
    df = pd.DataFrame(columns=['GSE Number', 'Subdirectory', 'Filename', 'Size'])
    
    # 填充DataFrame
    for path, size in file_info_list:
        gse_number, subdir_name, file_name = split_file_info(path)
        df = df.append({'GSE Number': gse_number, 'Subdirectory': subdir_name, 'Filename': file_name, 'Size': size}, ignore_index=True)
    
    return df

# 示例
file_info_list = df

df2 = create_dataframe(file_info_list)

# 显示数据框
print(df2)

# 如果需要，可以将数据框导出到CSV
# df.to_csv('file_info.csv', index=False)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
#输入结构
#从list获得每一行的GSE编号,生成url
def generate_urls_from_file(filename):
    """从包含GSE编号的文件中生成URL列表"""
    urls = []
    with open(filename, 'r') as file:
        for line in file:
            gse_id = line.strip()
            # 根据GSE编号生成URL
            url = f"https://ftp.ncbi.nlm.nih.gov/geo/series/{gse_id[:6]}nnn/{gse_id}/"
            urls.append(url)
    return urls

def extract_subdirectories(url):
    """从给定的URL中提取子目录名称"""
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    subdirs = []
    pre_tag = soup.find('pre')
    if pre_tag:  # 检查<pre>标签是否存在
        for a_tag in pre_tag.find_all('a'):
            href = a_tag.get('href')
            if href.endswith('/') and (href.startswith('m') or href.startswith('s')):
                subdir_name = href
                subdirs.append(subdir_name)
    return subdirs

def get_subdirectory_content(base_url, subdirs):
    """获取每个子目录的内容"""
    subdir_contents = {}
    for subdir in subdirs:
        suburl = base_url + '/' + subdir
        response = requests.get(suburl)
        subdir_contents[subdir] = response.text
    return subdir_contents
#以上会返回子目录的response的json格式信息, 再接一个新函数来挖掘信息

#获取完整信息
#此处获得的all file info的type是字典

def get_file_info_dataframe(subdir_contents,gse_id):
    all_file_info = {}
    for subdir, content in subdir_contents.items():
        soup = BeautifulSoup(content, 'html.parser')
        file_info = []
        for a_tag in soup.find('pre').find_all('a'):
            href = a_tag.get('href')
            if href.startswith('GSE'):
                file_name = href
                size = a_tag.find_next_sibling(string=True).strip().split()[-1]
                file_info.append((file_name, size))
        all_file_info[subdir] = file_info
#从字典中提取数据并转换为DataFrame
    data = []
    for subdir, files in all_file_info.items():
        for file_name, size in files:
            data.append([subdir.rstrip('/'), file_name, size])
    df = pd.DataFrame(data, columns=['Subdirectory', 'File Name', 'Size'])
    df['GSEID'] = gse_id
    return df
```

```python
# 初始化一个空的主DataFrame
main_df = pd.DataFrame()

# 从txt文件中读取每个GSE编号
urls = generate_urls_from_file('/Users/sunhao/DRaw_support/list4')

```

```python
# 遍历每个URL
for url in urls:
    # 提取子目录
    subdirs = extract_subdirectories(url)
    # 获取子目录内容
    subdir_contents = get_subdirectory_content(url, subdirs)
    # 获取文件信息并转换为DataFrame
    gse_id = url.split('/')[-2]  # 从URL中提取GSE编号
    df = get_file_info_dataframe(subdir_contents, gse_id)
    # 将新生成的df追加到主df中
    main_df = main_df.append(df)

# 重置主df的索引
main_df.reset_index(drop=True, inplace=True)
```
